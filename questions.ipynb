- check whether a number is even or odd

```python
num = int(input("Enter a number : "))

if num % 2 == 0:
   print(f'The {num} is even')
else:
   print(f'The {num} is odd')
```

- check whether a number is positive or negative

```python
num = int(input("Enter a number : "))

if num >= 0:
   print(f'{num} is positive')
elif num < 0:
   print(f'{num} is negative')
else:
   print("Invalid number")
```

- check whether a character entered by user is vowel or not


```python
char = input("Enter a character: ")
vowels = 'aeiouAEIOU'

if char in vowels:
    print(f'{char} is a vowel')
else:
    print(f'{char} is not a vowel')
```

- check whether a a triangle is an equilateral or iscoceles triangle


```python
a= int(input("Enter first side: "))
b = int(input("Enter second side: "))
c = int(input("Enter third side: "))

if a==b==c:
    print("Equilateral triangle")
elif a==b or a==c or b==c:
    print("Isosceles triangle")
else:
    print("Invalid triangle")    
```

- if the basec salary of an employee is 50000, HRA is 5% of basic salary, DA is 10% of basic salary. Calculate gross salary of an employee

```python
salary = int(input("Enter your salary: "))
HRA = 0.5 * salary
DA = 0.1 * salary

grosssal = HRA +  DA + salary
print(f'Gross Salary is {grosssal}')
```
