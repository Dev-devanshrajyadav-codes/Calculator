-Some string methods
- .lower() - str into lowercase
- .upper() - str into uppercase
- .capitalize() - str 1st letter into capital
- .title() - 1st letter of all words in capital
- .find() -return the first index of the occurence
- .replace() - replace with a word
- .split() - spilt the string
- .splitlines() - 

```python
str = "digiPodium class"
print(str.lower())

print(str.upper())

print(str.capitalize())
```
