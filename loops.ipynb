Write a program to calculate the factorial of a number using a for loop.

```python
num = input("Enter the number to find its factorial: ")
fact = 1

for i in range(1,int(num)+1):
    fact = fact * i
print("The factorial of ",num," is ",fact)
```

Write a program to print the Fibonacci series up to a given limit using a for loop.

```python
num = int(input("Enter the number to find its fibonacci series: "))
first = 0
second = 1

print("Fubonacci series: ")

for i in range(1, int(num) + 1):
    #print the first two numbers
    
    if i == 1:
        print(first)
    if i == 2:
        print(second)
    else:
        next =  first + second
        print(next)
        first = second
        second = next
              

```

Write a program to reverse a given string using a while loop.

```python
str = input("Enter the string to reverse: ")
reverse = ""
# it is the concatenation method to reverse a string 
for i in str:
    reverse = i + reverse
print("The reverse of the string is: ",reverse)
```

Write a program to print the ASCII values of all uppercase letters using a for loop.

```python
for ch in range(ord('A'), ord('Z') + 1):
    print(f"ASCII value of {chr(ch)} is {ch}")
```
