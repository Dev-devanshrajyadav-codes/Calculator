Prog - Wap to take name and phone number of 3 users and store them in dictionary

```python
dict = {}

for i in range(3): 
    name = input("Enter your name")
    phone_num = int(input("Enter your number"))

    dict[name] = phone_num

print(dict)
```

Typecasting - dict() method

```python
info = dict(name = "dev", age = 20, course = "BCA")
print(info)
```

```python
info.keys()
```

```python
info.items()
```

loops in dictionary

```python
for k,v in info.items():
    print(f'{k} - {v}')
```
