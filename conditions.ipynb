       - if + else
  - Multiple

```python
marks = int(input("Enter your marks in math: "))
if marks >= 90:
    print("A+")
if marks >= 75:
    print("A")
if marks >= 60:
    print("B")
if marks < 60:
    print("Fail")
```

```python
marks = int(input("Enter your marks in math: "))
if marks >= 90:
    print("A+")
elif marks >= 75:
    print("A")
elif marks >= 60:
    print("B")
else:
    
    print("Invalid marks")
```

 - In if statement - code executes atleast one for real, and if no cond is true gives blank 
 - In elif - if cond is true , breaks
