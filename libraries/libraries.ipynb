```python
from random import random , randint , choice , shuffle  
```

```python
print('random number between 0 and 1:', random()) #gives nearest value to 0 
print('random number between 0 and 10:',randint(1,10)) #will choose number from 0 to 10 randomly
print('random choice from list:',choice([2,5,4,6,7,8,9,2,3,5,3])) #will print any randam nbr from the made list 
my_list=["john","jane","jake","jill"] 
shuffle(my_list) #this command will shuffle the list every time it will run
print('shuffle list:', my_list)
```

```python
from datetime import datetime as dt #named as alais
```

```python
print("current date and time:", dt.now().date())
print("current date:",dt.now().date())
print("current time:",dt.now().time())
print("current year:",dt.now().year)
print("current month:",dt.now().month)
print("current day:",dt.now().day)
print("current hour:",dt.now().hour)
```

```python
print("fromatted date",dt.now().strftime("%d-%m-%y"))
print("fromatted time",dt.now().strftime("%H-%M-%S"))
print("current month name",dt.now().strftime("%b"))
```
