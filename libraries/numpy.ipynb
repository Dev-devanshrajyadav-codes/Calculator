### Array Properties

```python
a = np.array([[1,2,3,4],[4,5,6,7]])
print(a)  #always look to the starting brackets to know which dimenstion array 
```

```python
print(a.ndim) #number of dimensions
print(a.shape) #number of rows and coloumns
print(a.size) #number of elements
print(a.dtype) #data types of the elements
```

### Reshaping and flattening

```python
a.reshape(4,2) #reshape array
```

```python
a.flatten() #convert multi dimension array into 1D array
```

### Index and sliceing

```python
a = np.array([10,20,30,40,50])
```

```python
#to find array from index
print(a[0]) #frist element
print(a[1:4])
print(a[-1]) #last element
```

```python
b = np.array([[1,2,3],[4,5,6]])
```

```python
print(b[0]) #first row
print(b[0,1]) # 1 row 2nd column
```

### Mathematical Opretion - element wise opretion

```python
x= np.array([1,2,3,4,5])
y= np.array([4,5,6,7,8])
```

```python
print(x+y)
print(x-y)
print(x*y)
```
